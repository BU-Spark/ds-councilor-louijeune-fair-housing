```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('income-restricted-inventory-2021.csv')
grouped_data = data.groupby('Zip Code').agg({'Total Income-Restricted':'sum'}).reset_index()
sorted_data = grouped_data.sort_values('Total Income-Restricted', ascending=False)

plt.bar(sorted_data['Zip Code'], sorted_data['Total Income-Restricted'])

plt.title('Geographic Distribution of Affordable Housing by Zip Code')
plt.xlabel('Zip Code')
plt.ylabel('Affordable Housing')


plt.show()
print(sorted_data.head())
```

```python
df = pd.read_csv('income-restricted-inventory-2021.csv')
num_affordable_units = df['Total Income-Restricted'].sum()
num_non_affordable = df['TtlProjUnits'].sum() - num_affordable_units
print('Number of units enrolled in affordable housing in 2021:', num_affordable_units) #most updated we could find
print('Number of units not enrolled in affordable housing in 2021:', num_non_affordable)

```

```python
import pandas as pd

df = pd.read_csv('income-restricted-inventory-2021.csv')
df['Not Affordable'] = df['TtlProjUnits'] - df['Total Income-Restricted']
grouped = df.groupby('Neighborhood')['Not Affordable'].sum().reset_index()
grouped = grouped.sort_values('Not Affordable', ascending=False)
affordable_sum = df['Not Affordable'].sum()

print(f"In Boston these are the neighborhoods with the least Affordable Housing Offered {affordable_sum}.")
print(grouped)
```
