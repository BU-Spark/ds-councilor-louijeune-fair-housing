```python
shapefile_path = "../spark/census_tract_map/Census2020_Tracts.shp"

tracts2 = gpd.read_file(shapefile_path)
```

```python
joined = gpd.sjoin(tracts2, tracts1, op='intersects')

# randomly assign each tract in tracts2 to a neighborhood in tracts1
import random
neighborhoods = joined.groupby('index_right').groups.keys()
for neighborhood in neighborhoods:
    tracts = joined[joined['index_right'] == neighborhood]
    tracts2.loc[tracts.index, 'Neighbor'] = random.choice([tracts1.loc[neighborhood, 'Name']] * len(tracts))

tracts2
```

```python
#rejected rate for each race: rejected case of one race/overall rejection
df_rejected = df.loc[df["action_taken"] ==3]
df_rejected = df_rejected[['applicant_age', 'derived_dwelling_category', 'derived_race', 'income']]
df_rejected = df_rejected.loc[(df_rejected['applicant_age'] != "8888")]

rate_age_rejection =(df_rejected['applicant_age'].value_counts()/(df_rejected['applicant_age'].value_counts()+age_accepted['applicant_age'].value_counts()))*100
plt.figure(figsize=(8, 10))
fig, ax = plt.subplots()
rate_age_rejection.plot(kind='bar')
ax.set_xlabel('Ages')
ax.set_ylabel('Rate of Loan Rejection for Each Race(%)')
ax.set_title('Loan Rejection Rates by Each Race(Rejection/Sum of Application for Each Race)')

for index, value in enumerate(rate_age_rejection):
    value = round(value,2)
    plt.annotate(str(value)+"%", xy=(index, value), ha='center')

# plt.show()
plt.savefig('Loan Rejection Rates by Each Age(Rejection over Sum of Application for Each Race)')
```
